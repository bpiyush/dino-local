```python
def get_attention_maps(img, patch_size=patch_size, threshold=None):

    # make the image divisible by the patch size
    w, h = img.shape[1] - img.shape[1] % patch_size, img.shape[2] - img.shape[2] % patch_size
    img = img[:, :w, :h].unsqueeze(0)

    w_featmap = img.shape[-2] // patch_size
    h_featmap = img.shape[-1] // patch_size

    with torch.no_grad():
        attentions = model.get_last_selfattention(img.to(device))

    nh = attentions.shape[1] # number of head

    # we keep only the output patch attention
    attentions = attentions[0, :, 0, 1:].reshape(nh, -1)

    if threshold is not None:
        # we keep only a certain percentage of the mass
        val, idx = torch.sort(attentions)
        val /= torch.sum(val, dim=1, keepdim=True)
        cumval = torch.cumsum(val, dim=1)
        th_attn = cumval > (1 - args.threshold)
        idx2 = torch.argsort(idx)
        for head in range(nh):
            th_attn[head] = th_attn[head][idx2[head]]
        th_attn = th_attn.reshape(nh, w_featmap, h_featmap).float()
        # interpolate
        th_attn = nn.functional.interpolate(
            th_attn.unsqueeze(0), scale_factor=args.patch_size, mode="nearest",
        )[0].cpu().numpy()

    attentions = attentions.reshape(nh, w_featmap, h_featmap)
    attentions = torch.nn.functional.interpolate(
        attentions.unsqueeze(0), scale_factor=patch_size, mode="nearest",
    )[0].cpu().numpy()

    return attentions


def show_attention_maps(image, maps):
    maps = [su.visualize.tensor_to_heatmap(m) for m in maps]

    n_rows = 2
    n_cols = len(maps) // n_rows
    top_row = su.visualize.concat_images_with_border(maps[:n_cols])
    bot_row = su.visualize.concat_images_with_border(maps[n_cols:])
    maps_canvas = su.visualize.concat_images_vertically_with_border([top_row, bot_row])

    canvas = su.visualize.concat_images_with_border(
        [su.visualize.rescale_frame(image, 2.), maps_canvas]
    )
    display(canvas)
```

### Load images

```python
mean = np.array([0.485, 0.456, 0.406])
std = np.array([0.229, 0.224, 0.225])
imsize = 512

fwd_transform = pth_transforms.Compose([
    pth_transforms.CenterCrop(size=(270, 270)),
    pth_transforms.Resize(imsize),
    pth_transforms.ToTensor(),
    pth_transforms.Normalize(mean, std),
])
inv_transform = [
    pth_transforms.Normalize(
        mean=[0., 0., 0.], std=1./std,
    ),
    pth_transforms.Normalize(
        mean=-mean, std=[ 1., 1., 1. ],
    ),
    pth_transforms.ToPILImage(),
]
inv_transform = pth_transforms.Compose(inv_transform)

data_root = "/scratch/shared/beegfs/piyush/datasets"
data_path = os.path.join(data_root, "PouringIROS2019/resized_data_cut_clips_frames")
dataset = datasets.ImageFolder(data_path, transform=fwd_transform)

len(dataset)
```

```python
class DatasetWithBoundingBoxes(torch.utils.data.Dataset):
    """Same as before but now crops to bounding boxes."""

    def __init__(self, transform=None):
        self.samples = dataset.samples
        self.transform = transform

        metadata = torch.load(
            os.path.join(
                data_root,
                "PouringIROS2019/annotations/water_container_detections-v1.pt",
            ),
        )
        metadata = {
            k.split("/")[-1]: v for k, v in metadata.items()
        }
        self.boxes = [
            metadata["_".join((s[0].split("/")[8].split("_")[:2]))] for s in self.samples
        ]

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, i):
        sample = self.samples[i]
        img = PIL.Image.open(sample[0])
        box = self.boxes[i]
        img = img.crop(box)
        if self.transform is not None:
            img = self.transform(img)
        return img, sample[1]
```

```python
mean = np.array([0.485, 0.456, 0.406])
std = np.array([0.229, 0.224, 0.225])
imsize = 224

fwd_transform = pth_transforms.Compose([
    pth_transforms.Resize(imsize),
    pth_transforms.ToTensor(),
    pth_transforms.Normalize(mean, std),
])
inv_transform = [
    pth_transforms.Normalize(
        mean=[0., 0., 0.], std=1./std,
    ),
    pth_transforms.Normalize(
        mean=-mean, std=[ 1., 1., 1. ],
    ),
    pth_transforms.ToPILImage(),
]
inv_transform = pth_transforms.Compose(inv_transform)

dataset_bb = DatasetWithBoundingBoxes(transform=fwd_transform)
```
